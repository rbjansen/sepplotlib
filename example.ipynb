```python
import pandas as pd
import numpy as np
import sepplotlib as spl
```

```python
n = 100
np.random.seed(1962)

mock = pd.DataFrame(
    {
        "y_true": np.concatenate(
            [
                np.random.choice([0, 1], size=(int(n / 4),), p=p) 
                for p in ([0.9, 0.1], [0.6, 0.4], [0.4, 0.6], [0.1, 0.9])
            ]
        ),
        "y_pred_a": np.arange(0, 0.9, 0.9 / n) + np.random.uniform(0, 0.1, n),
        "y_pred_b": np.random.uniform(0, 1, n),
        "lab": [f"ID{i}" for i in range(0, n)]
    }
)
```

```python
spl.SeparationPlot(
    mock, "y_true", "y_pred_a", title="Example"
)
```

```python
spl.ModelCriticismPlot(
    mock, "y_true", "y_pred_a", "lab", title="Example", n_worst=5
)
```

```python
spl.BiseparationPlot(
    df=mock,
    x="y_pred_a",
    y="y_pred_b",
    obs="y_true",
    lab="lab",
    title="Example",
)
```
